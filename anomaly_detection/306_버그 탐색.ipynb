### Data EDA

- 데이터 특성 파악을 위한 초기 탐색

```python
# 기본 정보를 확인
for i in ['Id', 'groupId', 'matchId']:
    print(f'unique [{i}] count:', df[i].nunique())
```

```python
# box plot을 통한 분포 확인
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use(['dark_background'])

kills = df.copy()

kills['killsCategories'] = pd.cut(kills['kills'], [-1, 0, 2, 5, 10, 60], labels=['0_kills', '1-2_kills', '3-5_kills', '6-10_kills', '10+_kills'])
plt.figure(figsize=(15,8))
sns.boxplot(x='killsCategories', y='winPlacePerc', data=kills)
plt.show()
```

```python
# joint plot 으로 va1, va2 관계를 확인
# x축 : 승률, y : 도보이동거리
sns.jointplot(x='winPlacePerc', y='walkDistance', data=df, height=10, ratio=3, color="lime")
plt.show()
```

```python
# joint plot 으로 va1, va2 관계를 확인
# x축 : 승률, y : 차량이동거리
sns.jointplot(x='winPlacePerc', y='rideDistance', data=df, height=10, ratio=3, color="lime")
plt.show()
```

```python
# joint plot 으로 va1, va2 관계를 확인
# x축 : 승률, y : 헤드
sns.jointplot(x='winPlacePerc', y='headshotKills', data=df, height=10, ratio=3, color="lime")
plt.show()
```

### 이상 데이터 정의

- 가설을 수립하고, 데이터로 가볍게 검증

1. 자동 조준
2. 회복없이 Kill수
3. 100% 해드샷
4. 치료제 생성
5. 이동없이 무기 획득

```python
# 가설1 : 이동 거리가 낮은데 해드 비율이 높은가? (자동 조준)
df['_totalDistance'] = df['rideDistance'] + df['walkDistance'] + df['swimDistance']

df['_headshot_rate'] = df['headshotKills']/df['kills']
df['_headshot_rate'] = df['_headshot_rate'].fillna(0)

sns.scatterplot(x='_totalDistance', y='_headshot_rate', data=df, color="lime");
```

```python
# 가설2 : heal 없이 Kill 수가 많은 버그 유저가 존재할까?
sns.scatterplot(x='kills', y='heals', data=df, color="lime");
```

```python
# 가설3 : 100% Head만 하는 유저가 존재할까?
sns.scatterplot(x='_headshot_rate', y='kills', data=df, color="lime");
```

```python
# 가설4 : 치료제를 너무 많이 쓰는 유저가 존재할까? (치료제 생성)
sns.distplot(df['heals']);
```

```python
# 가설5 : 이동없이 무기를 획득하는 유저가 있을까?
sns.scatterplot(x='_totalDistance', y='weaponsAcquired', data=df, color="lime");
```

### 이상 탐지 모델링

```python
# 1. 자동 조준 - 분석
# _totalDistance, _headshot_rate 활용
# 10분위수(declie) : 내림 또는 오름차순으로 정렬했을 때 10등분으로 데이터를 나눔

# 이동거리 10% 이하면서 해드비율이 80% 이상인 데이터셋 추출
df_1 = df[(df['_totalDistance'] <= df['_totalDistance'].quantile(0.10)) & (df['_headshot_rate'] >= 0.8)]
print(len(df_1))
df_1.head()
```

```python
# 데이터가 너무 많으므로, kill 분포를 통해 이상한 데이터를 탐색
sns.distplot(df_1['kills']);
```

```python
df['_totalDistance'].quantile(0.10), df['_totalDistance'].quantile(0.90)
```

```python
# 버그 의심 유저 검출
df_1[df_1['kills'] > 6][['kills', '_totalDistance', '_headshot_rate']]
```

```python
# 2. 힐없이 다수의 킬
# 평균적인 Heal 횟수 출력
df[['heals', 'kills']].describe()
```

```python
# quantile(0.90) 로 구분하기에는 너무 적은 숫자의 kill 수
df['heals'].quantile(0.10), df['kills'].quantile(0.90)
```

```python
# heal은 0회, kill은 평균*3std 이상인 유저 확인
df['kills'].mean() + 3*df['kills'].std()
```

```python
# heal은 0회, kill은 평균*3std 이상인 유저 확인

df_2 = df[(df['heals'] <= df['heals'].quantile(0.10)) & (df['kills'] >= df['kills'].mean()+3*df['kills'].std())]
print(len(df_2))
df_2[['heals', 'kills']]
```

```python
# kill 수 분포 확인
sns.distplot(df_2['kills'])
```

```python
# 버그 의심 유저 검출
df_2[df_2['kills'] > 30][['kills', 'heals']].sort_values(by=['kills'])
```

```python
# 3. 해드 100% - 분석
df_3 = df[df['_headshot_rate'] == 1]
print(len(df_3))
df[df['_headshot_rate'] == 1].head()
```

```python
# kill 수 분포 확인
sns.distplot(df_3['kills']);
```

```python
df_3['kills'].describe()
```

```python
df_3['kills'].mean()+3 * df_3['kills'].std()
```

```python
# 버그 의심 유저 검출 (10명 이상 킬인데 모두 해드샷인 유저)
df_3[df_3['kills'] > 10][['kills', '_headshot_rate']]
```

```python
# 4. 힐 아이템 이상 사용 - 분석
# heal 아이템 사용횟수 분포 확인
sns.distplot(df['heals']);
```

```python
df['heals'].describe()
```

```python
# 버그 의심 유저 검출 (힐 50개 이상 사용)
df[df['heals']> 50]['heals']
```

```python
# 5. 움직임 없이 무기 획득 - 분석
# 기준치 확인 (이동거리 10%이하면서 무기획득이 90% 이상인 row)
df['_totalDistance'].quantile(0.10), df['weaponsAcquired'].quantile(0.90)
```

```python
# 이동거리 10% 이하면서 무기 획득 90% 이상인 유저
df_5 = df[(df['_totalDistance'] <= df['_totalDistance'].quantile(0.10)) & (df['weaponsAcquired'] >= df['weaponsAcquired'].quantile(0.90))]
print(len(df_5))
df_5.head()
```

```python
# 무기 획득에 대한 분포 확인
sns.distplot(df_5['weaponsAcquired']);
```

```python
# 버그 의심 유저 검출
df_5[df_5['weaponsAcquired'] > 50][['_totalDistance','weaponsAcquired']]
```
