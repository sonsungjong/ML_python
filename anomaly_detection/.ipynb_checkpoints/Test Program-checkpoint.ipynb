```python
import numpy as np
import matplotlib.pyplot as plt
import datetime

# Sample data
days = 5
hours = np.arange(0, 24 * days, 6)
temperature = np.random.randint(60, 100, size=len(hours))
humidity = np.random.randint(30, 80, size=len(hours))

# Plot
fig, ax1 = plt.subplots()

# Temperature
color = 'tab:red'
ax1.set_xlabel('Time (days)')
ax1.set_ylabel('Temperature (°F)', color=color)
ax1.plot(hours, temperature, 'o', color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Humidity
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Humidity (%)', color=color)
ax2.plot(hours, humidity, '^', color=color)
ax2.tick_params(axis='y', labelcolor=color)

# X-axis ticks
ticks = np.arange(0, 24 * days, 24)
labels = [str(datetime.timedelta(hours=int(hours))) for hours in ticks]
ax1.set_xticks(ticks)
ax1.set_xticklabels(labels)


# Display plot
fig.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 더미 데이터 생성
x = np.linspace(-40, 100, 15)  # -40에서 100까지의 x값
y_temperature = np.random.randint(0, 50, len(x))  # 온도 데이터 (섭씨)
y_humidity = np.random.randint(0, 100, len(x))  # 습도 데이터

# 그래프 생성
fig, ax = plt.subplots()

ax.plot(x, y_temperature, '-o', label='Temperature (°C)', markersize=8)
ax.plot(x, y_humidity, '-^', label='Humidity (%)', markersize=8)

# 축 레이블 설정
ax.set_xlabel('X-axis')
ax.set_ylabel('Values')

# 범례 추가
ax.legend(loc='best')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime

# 데이터
temperature_data = np.array([-32, -20, 0, 20, 40, 50, 40, 30])
humidity_data = np.array([20, 40, 70, 60, 40, 50, 82, 55])

# X축 날짜 데이터 생성
date1 = datetime.datetime(2023, 4, 27)
date2 = datetime.datetime(2023, 4, 28)
delta = (date2 - date1) / 7
x_dates = [date1 + i * delta for i in range(8)]

# 그래프 생성
fig, ax = plt.subplots()

ax.plot(x_dates, temperature_data, '-o', label='Temperature (°C)', markersize=8)
ax.plot(x_dates, humidity_data, '-^', label='Humidity (%)', markersize=8)

# 축 레이블 설정
ax.set_xlabel('Date')
ax.set_ylabel('Values')

# X축 날짜 형식 지정
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

# 범례 추가
ax.legend(loc='best')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime

# 데이터
temperature_data = np.array([-32, -20, 0, 20, 40, 50, 40, 30, -32, -20, 0, 20, 40, 50, 40, 30, -32, -20, 0, 20, 40, 50, 40, 30])
humidity_data = np.array([20, 40, 70, 60, 40, 50, 82, 55, 20, 40, 70, 60, 40, 50, 82, 55, 20, 40, 70, 60, 40, 50, 82, 55])

# X축 날짜 데이터 생성
date1 = datetime.datetime(2023, 4, 27)
date2 = datetime.datetime(2023, 4, 28)
date3 = datetime.datetime(2023, 4, 29)
date4 = datetime.datetime(2023, 4, 30)

delta = (date2 - date1) / 6
x_dates = [date1 + i * delta for i in range(6)] + [date2 + i * delta for i in range(6)] + [date3 + i * delta for i in range(6)] + [date4 + i * delta for i in range(6)]

# 그래프 생성
fig, ax = plt.subplots()

ax.plot(x_dates, temperature_data, '-o',  markersize=8)
ax.plot(x_dates, humidity_data, '-^', markersize=8)

# 축 레이블 설정
ax.set_xlabel('Date')
ax.set_ylabel('Values')

# X축 날짜 형식 지정
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

# 중복되는 X축 레이블 제거
for label in ax.xaxis.get_ticklabels()[1:23:2]:
    label.set_visible(False)

# 구분선 추가
for i in range(4, 24, 4):
    ax.axvline(x=x_dates[i], color='gray', linestyle='--', alpha=0.5)

# 범례 추가
ax.legend(loc='best')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime

# 데이터
temperature_data = np.array([-32, -20, 0, 20, 40, 50, 40, 30, -32, -20, 0, 20, 40, 50, 40, 30, -32, -20, 0, 20, 40, 50, 40, 30])
humidity_data = np.array([20, 40, 70, 60, 40, 50, 82, 55, 20, 40, 70, 60, 40, 50, 82, 55, 20, 40, 70, 60, 40, 50, 82, 55])

# 날짜
date1 = datetime.datetime(2023, 4, 27)
date2 = datetime.datetime(2023, 4, 28)
date3 = datetime.datetime(2023, 4, 29)
date4 = datetime.datetime(2023, 4, 30)

# 시간
time_stamps = []
for date in [date1, date2, date3, date4]:
    for hour in [0, 6, 12, 18]:
        time_stamps.append(date + datetime.timedelta(hours=hour))

# 그래프 생성
fig, ax1 = plt.subplots()

ax1.plot(time_stamps, temperature_data, 'o-', label='Temperature', linewidth=2)
ax1.plot(time_stamps, humidity_data, 'v-', label='Humidity', linewidth=2)

# Y축 20도 단위 점선
ax1.yaxis.set_ticks(np.arange(-40, 120, 20))
ax1.grid(axis='y', linestyle='--', linewidth=0.5)

# X축 레이블
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(time_stamps[::4], rotation=45)

# 구분선 추가
for i in range(1, 4):
    ax1.axvline(x=time_stamps[i*4], color='gray', linestyle='--', linewidth=0.5)

# 범례 추가
ax1.legend(loc='best')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime

# 데이터
temperature_data = np.array([-32, -20, 0, 20, 40, 50, 40, 30, -32, -20, 0, 20, 40, 50, 40, 30])
humidity_data = np.array([20, 40, 70, 60, 40, 50, 82, 55, 20, 40, 70, 60, 40, 50, 82, 55])

# 날짜
date1 = datetime.datetime(2023, 4, 27)
date2 = datetime.datetime(2023, 4, 28)
date3 = datetime.datetime(2023, 4, 29)
date4 = datetime.datetime(2023, 4, 30)

# 시간
time_stamps = []
for date in [date1, date2, date3, date4]:
    for hour in [0, 6, 12, 18]:
        time_stamps.append(date + datetime.timedelta(hours=hour))

# 그래프 생성
fig, ax1 = plt.subplots()

# 온도 데이터 표시
temperature_line, = ax1.plot(time_stamps, temperature_data, 'o-', label='Temperature', linewidth=2)
for x, y in zip(time_stamps, temperature_data):
    ax1.annotate(f'{y}', xy=(x, y), xytext=(-5, 7), textcoords='offset points', fontsize=8, color=temperature_line.get_color())

# 습도 데이터 표시
humidity_line, = ax1.plot(time_stamps, humidity_data, 'v-', label='Humidity', linewidth=2)
for x, y in zip(time_stamps, humidity_data):
    ax1.annotate(f'{y}', xy=(x, y), xytext=(-5, -10), textcoords='offset points', fontsize=8, color=humidity_line.get_color())

# Y축 20도 단위 점선
ax1.yaxis.set_ticks(np.arange(-40, 120, 20))
ax1.grid(axis='y', linestyle='--', linewidth=0.5)

# X축 레이블
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(time_stamps[::4], rotation=45)

# 구분선 추가
for i in range(1, 4):
    ax1.axvline(x=time_stamps[i*4], color='gray', linestyle='--', linewidth=0.5)

# 범례 추가
ax1.legend(loc='best')

plt.show()
```
