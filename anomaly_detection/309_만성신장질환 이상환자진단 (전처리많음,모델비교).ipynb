### 데이터 전처리

- 수집된 데이터의 기본 정보들을 확인
1. Data shape
2. Data info(자료형)
3. Null 확인

```python
# 데이터 테이블 확인 (400 row, 26 col)
# 총 데이터 400 튜플이고 각 26 필드 보유
print('df :', df.shape)
```

```python
# 자료형 확인
df.info()
```

```python
# index 역할의 id는 Drop
df.drop('id', axis=1, inplace=True)
```

```python
# 약어로 되어있는 Column명을 변경
df.columns = ['age', 'blood_pressure', 'specific_gravity', 'albumin', 'sugar', 'red_blood_cells', 'pus_cell', 
              'pus_cell_clumps', 'bacteria', 'blood_glucose_random', 'blood_urea', 'serum_creatinine', 'sodium',
             'potassium', 'haemoglobin', 'packed_cell_volume', 'white_blood_cell_count', 'red_blood_cell_count',
             'hypertension', 'diabetes_mellitus', 'coronary_artery_disease', 'appetite', 'peda_edema', 'aanemia', 'class']
```

```python
# 숫자값인데 자료형이 object 인 필드를 변경
# errors='coerce' : 만약 숫자로 변경할 수 없는 데이터면 NaN으로 반환

df['packed_cell_volume'] = pd.to_numeric(df['packed_cell_volume'], errors='coerce')
df['white_blood_cell_count'] = pd.to_numeric(df['white_blood_cell_count'], errors='coerce')
df['red_blood_cell_count'] = pd.to_numeric(df['red_blood_cell_count'], errors='coerce')
```

```python
df[['packed_cell_volume', 'white_blood_cell_count', 'red_blood_cell_count']]
```

```python
df[['packed_cell_volume', 'white_blood_cell_count', 'red_blood_cell_count']].info()
```

```python
# Null 값 확인
print(df.isnull().sum())
```

```python
# 숫자형과 범주형을 분할
numerical_list = []
categorical_list = []

for i in df.columns:
    if df[i].dtypes == 'O':
        categorical_list.append(i)
    else:
        numerical_list.append(i)
        
print('categorical_list :', categorical_list)
print('numerical_list :', numerical_list)
```

```python
# 문자열 컬럼 unique 값 확인
for col in categorical_list:
    print(f'{col} has {df[col].unique()} values \n')
```

```python
# replace 를 통해서 부적절한 값은 변경

df['diabetes_mellitus'].replace(to_replace = {'\tno':'no', '\tyes':'yes', ' yes':'yes'}, inplace=True)
df['coronary_artery_disease'] = df['coronary_artery_disease'].replace(to_replace='\tno', value='no')
df['class'] = df['class'].replace(to_replace = {'ckd\t':'ckd', 'notckd': 'not ckd'})
```

```python
df['class'].value_counts()
```

```python
# 목표 Target 변수 변환 (0 또는 1로 값을 매핑하고, 숫자형으로 변경한다)
df['class'] = df['class'].map({'ckd':0, 'not ckd' : 1})
df['class'] = pd.to_numeric(df['class'], errors='coerce')
```

```python
# 오류 데이터 정상 변환 확인
cols = ['diabetes_mellitus', 'coronary_artery_disease', 'class']

for col in cols:
    print(f'{col} has {df[col].unique()} values \n')
```

```python
# Null 값 처리를 위한 2가지 방법
# 1. 랜덤으로 샘플링해서 Null에 채워넣음
def random_value_imputation(feature):
    random_sample = df[feature].dropna().sample(df[feature].isna().sum())
    random_sample.index = df[df[feature].isnull()].index
    df.loc[df[feature].isnull(), feature] = random_sample

# 2. 최빈값 mode를 대입 (null 비율이 적을 경우)
def impute_mode(feature):
    mode = df[feature].mode()[0]
    df[feature] = df[feature].fillna(mode)
```

```python
df[numerical_list].isnull().sum()
```

```python
# null 값이 아닌 값들 중 Random Sampling한 값을 부여
for col in numerical_list:
    random_value_imputation(col)
```

```python
df[numerical_list].isnull().sum()
```

```python
# null 값이 데이터의 많은 비율을 차지할 때, mode(최빈값)를 사용하면 데이터를 변질시킬 수 있기 때문에
# null 비율이 높은 object 필드는 랜덤 샘플링으로 진행
random_value_imputation('red_blood_cells')
random_value_imputation('pus_cell')

# 나머지 categorical 필드에 대해서는 최빈값mode 입력
for col in categorical_list:
    impute_mode(col)
```
