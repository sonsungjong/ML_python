## Features 항목

항목의 종류
- 범주형 항목 (Categorical Features)
범주형 항목은 범주형 변수로 된 항목으로 범주형 변수는 둘 이상의 결과 요소가 있는 변수이며 해당 기능의 각 값을 범주별로 분류할 수 있습니다.

ex) 성별은 두 가지 범주 (남성, 여성)의 범주형 변수입니다.

- 이산형 변수 (discrete variable)
범주형 변수 중 하나로 명목 변수(norminal variable)라고도 합니다.
데이터 셋에서 명목항목 : Sex, Embark
Name, Ticket 컬럼도 명목항목으로 변환해서 사용해야 합니다.

- Ordinal Variable (순위 변수)
순위 변수는 범주형 변수 중 하나이지만 차이점이 있습니다. 값 사이의 상대 순서(서열) 또는 정렬이 가능합니다.
데이터 셋에서 순위 항목은 PClass 입니다.
Cabin 컬럼은 순위 범주로 변환해서 사용해야 합니다.

- 연속형 항목 (Continuous Features)
서로 연속된 값을 가진 변수를 가진 항목이며 데이터 셋에서는 Age가 대표적입니다.
Age, SipSp, Parch, Fare는 interval variable로 만들어서 여기에 적용해야 합니다.

1. Variable 정의
2. survival 생존여부
3. pclass 선실 등급
4. sex 성별
5. age 나이
6. sibsp 가족유무 (형제자매/배우자)
7. parch 가족유무 (부모/자식)
8. ticket 표번호
9. fare 요금
10. cabin 선실번호
11. embarked 승선한 항구

### Ensemble
- 앙상블은 여러개의 Learner를 합하여 더 좋은 결과를 내는 기법
- Boosting 과 Bagging 이 있다.

### Boosting
- 각 Weak Classifier는 신뢰도가 떨어지나 합하면서 신뢰도를 높이는 방식

1. 목소리가 크니 경상도 사람이다
2. 자이언츠 유니폼을 입었으니 경상도 사람이다
3. 단어 선택을 보니 경상도 사람이다
4. 롯데 경기에 가는 것을 보니 경상도 사람이다

==> 4가지를 조합하면 경상도 사람이라는 것은 상당히 정확해진다

==> 좋은 결과를 낼 때까지 계속 연결한다.

### XGB
- XGBoosting은 Gradient Boosting을 발전시킨 버전으로 속도와 효율성을 증진시킴

### SVC
- 마진을 줘서 분류 실수를 어느 정도 용인

### AdaBoost
