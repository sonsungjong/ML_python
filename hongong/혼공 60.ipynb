```python
# 특성을 x축에, 타겟을 y축에 놓는다

import matplotlib.pyplot as plt

# scatter() : 산점도
plt.scatter(perch_length, perch_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 트레인셋과 테스트셋을 나눈다 (모델 신뢰성)

from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight, random_state=42)
```

```python
# 원본 데이터
test_array = np.array([1,2,3,4])
print(test_array.shape)

# 머신러닝 훈련을 위해 2차원배열 형태로 바꾼다 reshape()
test_array = test_array.reshape(2,2)
print(test_array.shape)
```

```python
# 자료형을 2차원배열로 바꾼다 (머신러닝 훈련을 위해)

train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)
print(train_input.shape, test_input.shape)
```

```python
from sklearn.neighbors import KNeighborsRegressor

knr = KNeighborsRegressor()

# KNN 회귀모델 훈련
knr.fit(train_input, train_target)
```

```python
# 테스트 점수 (정답을 맞힌 개수의 비율)
print(knr.score(test_input, test_target))
```

```python
from sklearn.metrics import mean_absolute_error

# 테스트 세트에 대한 예측을 만든다
test_prediction = knr.predict(test_input)

# 평균 절댓값 오차를 계산
mae = mean_absolute_error(test_target, test_prediction)
print(mae)
```

```python
# 테스트 세트의 점수가 훈련 세트의 점수보다 지나치게 낮다 : 과대적합
# 훈련 세트의 점수가 테스트 세트 점수보다 낮다(또는 둘다 너무 낮다) : 과소적합 (훈련세트와 테스트 세트의 양이 너무 적다)
print(knr.score(train_input, train_target))
```

```python
# 과소적합은 모델을 좀더 복잡하게 만들어 보정한다 (하이퍼 파라미터 조정)
knr.n_neighbors = 3

# 모델 재훈련
knr.fit(train_input, train_target)

# 훈련세트의 점수
print(knr.score(train_input, train_target))
```

```python
# 테스트세트의 점수
print(knr.score(test_input, test_target))
```
