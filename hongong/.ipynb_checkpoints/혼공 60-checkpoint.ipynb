```python
import numpy as np

perch_length = np.array([8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 21.0,
       21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 22.5, 22.7,
       23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 27.3, 27.5, 27.5,
       27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 36.5, 36.0, 37.0, 37.0,
       39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 40.0, 42.0, 43.0, 43.0, 43.5,
       44.0])
perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])
```

```python
# 데이터를 훈련세트와 테스트 세트로 나눈다
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(perch_length, perch_weight, random_state=42)
```

```python
# 2차 방정식 그래프로 변경
train_poly = np.column_stack((train_input **2, train_input))
test_poly = np.column_stack((test_input **2, test_input))
```

```python
# 새로 생성한 데이터셋 확인
print(train_poly.shape, test_poly.shape)
```

```python
lr = LinearRegression()
lr.fit(train_poly, train_target)

print(lr.predict([[50**2, 50]]))
```

```python
# [2차항계수, 1차항계수], 절편 (1.01*길이^2 -21.6*길이 +116.1)
print(lr.coef_, lr.intercept_)
```

```python
# 다항 회귀
point = np.arange(15, 50)

plt.scatter(train_input, train_target)

plt.plot(point, 1.01*point**2 -21.6*point +116.05)

predict_50 = lr.predict([[50**2, 50]])

# 50cm 농어 데이터
plt.scatter(50, predict_50, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
print(lr.score(train_poly, train_target))    # 훈련세트 점수
print(lr.score(test_poly, test_target))      # 테스트세트 점수
```

테스트세트가 훈련세트보다 점수가 높기 때문에 과소적합

과소적합은 모델을 좀더 복잡하게 만들어 해결한다

feature를 증가시키기 (길이 -> 길이, 높이, 두께)

## 특성 공학 (Feature Engineering)

기존 특성을 사용해 새로운 특성을 뽑아내는 작업

```python
import pandas as pd

# 특성 : 길이, 높이, 두께
df = pd.read_csv('https://bit.ly/perch_csv_data')
perch_full = df.to_numpy()
print(perch_full)
```

```python
import numpy as np

# Target 데이터 (무게)
perch_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0])
```

```python
# 훈련세트와 테스트세트로 나눈다
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(perch_full, perch_weight, random_state=42)
```

```python
# 새로운 특성을 만들거나 전처리 하기 위한 트랜스포머 사용
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures()
poly.fit([[2,3]])                    # 학습
print(poly.transform([[2,3]]))       # 변환
```

트랜스포머 결과

[[2,3]]  --> [[1,2,3,4,6,9]]

```python
poly = PolynomialFeatures(include_bias=False)    # 절편 제외
poly.fit([[2,3]])
poly.transform([[2,3]])
```

```python
poly = PolynomialFeatures(include_bias=False)
poly.fit(train_input)
train_poly = poly.transform(train_input)
train_poly.shape
```

```python
poly.get_feature_names_out()
```

```python
test_poly = poly.transform(test_input)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_poly, train_target)
```

```python
print(lr.score(train_poly, train_target))
print(lr.score(test_poly, test_target))
```

```python
# 2차항보다 더 크게 차수를 키워보면?
poly = PolynomialFeatures(degree=5, include_bias=False)          # 5 차수
poly.fit(train_input)
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)

train_poly.shape       # 특성 수 55개
```

```python
# 5제곱 후 훈련
lr.fit(train_poly, train_target)
print(lr.score(train_poly, train_target))
```

```python
# 계수를 증가시켰더니 과대적합 (너무 복잡, 데이터 수에 비해 모델이 너무 복잡)
lr.score(test_poly, test_target)
```

## 규제

모델이 너무 과도하게 학습하지 못하도록 방해

- Ridge
- Lasso

```python
# 규제하기 전에 먼저 정규화를 진행한다
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_poly)
train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)
```

```python
# 릿지 회귀
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
```

```python
import matplotlib.pyplot as plt

train_score = []
test_score = []
```

```python
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]

for alpha in alpha_list:
    ridge = Ridge(alpha=alpha)
    ridge.fit(train_scaled, train_target)
    train_score.append(ridge.score(train_scaled, train_target))
    test_score.append(ridge.score(test_scaled, test_target))
```

```python
plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()

# 최적값은 -1
```

```python
# 라쏘 회귀
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(train_scaled, train_target)
```

```python
print(lasso.score(train_scaled, train_target))
print(lasso.score(test_scaled, test_target))
```

```python
# 알파값을 바꾸어보며 계산
train_score = []
test_score = []
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
    lasso = Lasso(alpha=alpha, max_iter=10000)
    lasso.fit(train_scaled, train_target)
    train_score.append(lasso.score(train_scaled, train_target))
    test_score.append(lasso.score(test_scaled, test_target))
```

```python
plt.plot(np.log10(alpha_list), train_score)
plt.plot(np.log10(alpha_list), test_score)
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()
```

```python
# 10의 1승이 가장 근접
lasso = Lasso(alpha=10)
lasso.fit(train_scaled, train_target)
```

```python
print(lasso.score(train_scaled, train_target))
print(lasso.score(test_scaled, test_target))
```

```python
# 계수가 0인 것은 몇개인가?
print(np.sum(lasso.coef_ == 0))
```
