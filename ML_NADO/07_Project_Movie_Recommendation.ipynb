```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('tmdb_5000_credits.csv')
df2 = pd.read_csv('tmdb_5000_movies.csv')
```

```python
# df1의 컬럼명을 변경 (movie_id -> id)
df1.columns = ['id', 'title', 'cast', 'crew']
df1.columns
```

```python
# title 컬럼 제외
df1[['id', 'cast', 'crew']]
```

```python
# id라는 컬럼을 기준으로 df2에 df1을 병합
df2 = df2.merge(df1[['id', 'cast', 'crew']], on='id')
df2.head(3)
```

```python
# 영화1 평점이 10/10 -> 5명 평가
# 영화2 평점이 8/10 -> 500명 평가 (평점은 낮으나 신뢰도가 더 높음 -> 가중치를 적용해야함)
```

```python
C = df2['vote_average'].mean()
C
```

```python
# 상위 10퍼센트에 대한 데이터
m = df2['vote_count'].quantile(0.9)
m
```

```python
# df2 원본을 남겨놓기 위해 카피본을 생성 (vote_count 컬럼 중 vote_count 상위 10퍼 데이터만, 평가수가 적은 것들은 제외)
q_movies = df2.copy().loc[df2['vote_count'] >= m]
```

```python
q_movies['vote_count'].sort_values()
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (v / (v+m) * R) + (m / (m+v)*C)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
q_movies.head(3)
```

```python
# 점수를 기준으로 내림차순 정렬
q_movies = q_movies.sort_values('score',ascending=False)
```

```python
# 해당 컬럼만 출력
q_movies[['title','vote_count','vote_average','score']].head(10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# 무의미한 단어 제외 (a, an, of, ...)
tfidf = TfidfVectorizer(stop_words='english')
```

```python
# 해당 단어들이 제외됨
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
ENGLISH_STOP_WORDS
```

```python
# 데이터 전처리
# null 값이 있으면 빈문자로 교체
df2['overview'] = df2['overview'].fillna('')
```

```python
tfidf_matrix = tfidf.fit_transform(df2['overview'])
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim
```

```python
indices = pd.Series(df2.index, index=df2['title']).drop_duplicates()
indices
```

```python
# 영화 제목을 통해서 전체 데이터 기준 그 영화의 index 값을 얻기
indices['The Dark Knight Rises']
```

```python
df2.iloc[3]
```

```python
# 영화의 제목을 입력받으면 코사인 유사도를 통해서 가장 유사도가 높은 상위 10개의 영화 목록 반환
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    movie_indices = [i[0] for i in sim_scores]
    recommend_movie10 = df2['title'].iloc[movie_indices]
    return recommend_movie10
```

```python
get_recommendations('Avengers: Age of Ultron')
```

```python
# 감독 정보 추출
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
# Director에 해당하는 crew를 넣음
# Director 컬럼 추가
df2['director'] = df2['crew'].apply(get_director)
df2['director']
```

```python
# 처음 3개의 데이터 중 name에 해당하는 value 만 추출
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        if len(names) > 3:
            names = names[:3]
        return names
    return []
```

```python
features = ['cast','keywords','genres']
for feature in features:
    df2[feature] = df2[feature].apply(get_list)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(' ','')) for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x.replace(' ' , ''))
        else:
            return ''
```

```python
features = ['cast','keywords','director', 'genres']
for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

```python
df2[['title', 'cast', 'director', 'keywords', 'genres']].head(3)
```

```python
def create_soup(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + ' ' + x['director'] + ' ' + ' '.join(x['genres'])
df2['soup'] = df2.apply(create_soup, axis=1)
df2['soup']
```

```python
df2['overview']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(stop_words='english')
count_matrix = count.fit_transform(df2['soup'])
count_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# 코사인 유사도
cosine_sim2 = cosine_similarity(count_matrix, count_matrix)
cosine_sim2
```
