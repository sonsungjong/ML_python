```python
# 비지도학습은 y 가 필요 없음
# 독립변수 모든 데이터를 가져옴
X = dataset.iloc[:, :].values
```

```python
# x축 : hour , y축 : score / 산포도 그래프
plt.scatter(X[:,0], X[:,1])

plt.title('Score by hours')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

### 데이터 시각화 (축 범위 동일)

```python
# x축 : hour , y축 : score / 산포도 그래프
plt.scatter(X[:,0], X[:,1])

plt.title('Score by hours')
plt.xlabel('hours')
plt.xlim(0, 100)          # x축 범위 0~100
plt.ylabel('score')
plt.ylim(0, 100)           # y축 범위 0~100
plt.show()
# 실제 클러스터 거리 계산은 x축과 y축을 동일하게 범위 잡기 떄문에
```

### 피처 스케일링 (Feature Scaling)

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X = sc.fit_transform(X)
X
```

### 데이터 시각화 (스케일링된 데이터)

```python
# 스케일링된 데이터 사용
plt.figure(figsize=(5,5))
plt.scatter(X[:,0], X[:,1])

plt.title('Score by hours')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

## 최적의 클러스터(K) 값 찾기 (엘보우 방식 Elbow Method)

```python
from sklearn.cluster import KMeans
```

```python
inertia_list = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(X)    # 모델 생성 (학습)
    inertia_list.append(kmeans.inertia_)   # 각 지점으로부터 클러스터의 중심(centroid) 까지의 거리의 제곱의 합
```

```python
plt.plot(range(1, 11), inertia_list)
plt.title('Elbow Method')
plt.xlabel('n_clusters')
plt.ylabel('inertia')
plt.show()
```

### 최적의 K 값으로 KMeans 학습 (K=클러스터 갯수)

```python
# 최적의 K 값 (클러스터 갯수)
K = 4
```

```python
kmeans = KMeans(n_clusters=K, random_state=0)
# kmeans.fit(X)
y_kmeans = kmeans.fit_predict(X)    # 모델 생성(학습) 후 클러스터 위치 반환까지
```

```python
y_kmeans
```

### 데이터 시각화 (최적의 K)

```python
# 클러스터의 중심점 (centroid) 좌표
centers = kmeans.cluster_centers_
centers
```

```python
for cluster in range(K):
    plt.scatter(X[y_kmeans == cluster, 0], X[y_kmeans == cluster, 1], edgecolor='black', s=100)
    plt.scatter(centers[cluster, 0], centers[cluster, 1], color='yellow', edgecolor='black', s=300, marker='s')
    plt.text(centers[cluster, 0], centers[cluster, 1], cluster, va='center', ha='center')
    
plt.title('Score by hours')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

### 데이터 시각화 (스케일링 원복)

```python
# Feature Scaling 된 데이터를 다시 원복
X_origin = sc.inverse_transform(X)
centers_origin = sc.inverse_transform(centers)
```

```python
X_origin
```

```python
centers_origin
```

```python
for cluster in range(K):
    plt.scatter(X_origin[y_kmeans == cluster, 0], X_origin[y_kmeans == cluster, 1], edgecolor='black', s=100)
    plt.scatter(centers_origin[cluster, 0], centers_origin[cluster, 1], color='yellow', edgecolor='black', s=300, marker='s')
    plt.text(centers_origin[cluster, 0], centers_origin[cluster, 1], cluster, va='center', ha='center')
    
plt.title('Score by hours')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```
