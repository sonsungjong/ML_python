# 3. Polynomial Regression (지도학습 - 회귀)

## 공부 시간에 따른 시험 점수 (우등생)

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('PolynomialRegressionData.csv')
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

### 3-1. 단순 선형 회귀

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()

# 학습하여 모델 생성
reg.fit(X, y)
```

### 데이터 시각화 (전체)

```python
# 점 산점도 그래프
plt.scatter(X, y, color='blue')
# 선 그래프 (모델)
plt.plot(X, reg.predict(X), color='green')

plt.title('Score by hours (genius)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
# 전체 데이터를 통한 모델 평가
reg.score(X, y)
```

## 3-2 다항 회귀 (Polynomial Regression)

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
# 2차 다항식
# poly_reg = PolynomialFeatures(degree=2)
# 4차 다항식
poly_reg = PolynomialFeatures(degree=4)

# degree를 너무 높은 차수로 주면 과대적합이 나와서 오히려 부정확할 수 있음
X_poly = poly_reg.fit_transform(X)
X_poly[:5]
```

```python
# 선형 회귀
X[:5]
```

```python
lin_reg = LinearRegression()
# 변환된 X와 y로 모델 생성 (학습)
lin_reg.fit(X_poly, y)
```

### 데이터 시각화 (변환된 X와 y)

```python
plt.scatter(X, y, color='blue')
plt.plot(X, lin_reg.predict(poly_reg.fit_transform(X)), color='green')

plt.title('Score by hours (genius)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

```python
# X의 최소값에서 최대값까지의 범위를 0.1 단위로 잘라서 데이터 생성
X_range = np.arange(min(X), max(X), 0.1)
X_range
```

```python
# 2차원 배열 형태로 변경
# row 갯수 자동계산, column 갯수 1개
X_range = X_range.reshape(-1, 1)
X_range
```

```python
plt.scatter(X, y, color='blue')
# 선을 부드럽게 만들기 위해 X_range 사용
plt.plot(X_range, lin_reg.predict(poly_reg.fit_transform(X_range)), color='green')

plt.title('Score by hours (genius)')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
```

# 공부 시간에 따른 시험 성적 예측

```python
# 2시간 공부 했을 때 선형 회귀 모델 예측
reg.predict([[2]])
```

```python
# 시간 공부 했을 때 다항 회귀 모델 예측
lin_reg.predict(poly_reg.fit_transform([[2]]))
```

```python
lin_reg.score(X_poly, y)
```
